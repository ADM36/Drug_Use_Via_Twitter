```python
import re
import csv
import pandas as pd
import numpy as np
from ekphrasis.classes.preprocessor import TextPreProcessor
from ekphrasis.dicts.emoticons import emoticons
from ekphrasis.classes.segmenter import Segmenter
from nltk.tokenize import TweetTokenizer
from clean_tweets import clean_tweets
from replace import replace_utf
```

#### 1) Load in the train and test data

```python
train = pd.DataFrame()

# super inefficient but o well
tweet_id = []
adr_mention = []
tweet_text = []

with open('../train-split.tsv') as f:
    next(f) # skip header
    for line in f:
        line_sep = line.split('\t')

        tweet_id.append(line_sep[0])
        adr_mention.append(line_sep[1])
        tweet_text.append(line_sep[2].strip())

train['tweet_id'] = tweet_id
train['adr_mention'] = adr_mention
train['tweet_text'] = tweet_text

train.head(10)
```

```python
test = pd.DataFrame()

tweet_id = []
adr_mention = []
tweet_text = []

with open('../test-split.tsv') as f:
    next(f) # skip header
    for line in f:
        line_sep = line.split('\t')

        tweet_id.append(line_sep[0])
        adr_mention.append(line_sep[1])
        tweet_text.append(line_sep[2].strip())

test['tweet_id'] = tweet_id
test['adr_mention'] = adr_mention
test['tweet_text'] = tweet_text

test.head(10)
```

#### 2) Call replace_utf function to remove utf codes 
> Takes in a df and returns a df

```python
train = replace_utf(train)
train.head(10)
```

```python
test = replace_utf(test)
test.head(10)
```

#### 3) Call clean_tweets to do the rest of the cleaning and processing
> Takes in a df and returns a df

```python
train = clean_tweets(train)
train.head(10)
```

```python
test = clean_tweets(test)
test.head(10)
```

```python
export_csv = test.to_csv (r'test_25k.tsv', index = None, header = None, sep = '\t')
```
